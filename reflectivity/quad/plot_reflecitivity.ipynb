```python
import numpy as np
import matplotlib.pyplot as plt
import glob

%matplotlib nbagg
```

```python
def cos(theta):
    return np.cos(theta*np.pi/180)
```

```python
files = glob.glob('*.txt')

fig, ax = plt.subplots()
for file in files:
    M = np.loadtxt(file)
    ax.semilogy(M[:,0]-90, M[:,1], label=file[:-4])

ax.set_xlabel(r'$\theta$ ($^\circ$)')
ax.set_ylabel('I')
ax.legend()
```

```python
def get_peak(file):
    M = np.loadtxt(file)
    peakregioninds = np.logical_and(M[:,0]>(90+19), M[:,0]<(90+21))
    M_region = M[peakregioninds,:]
    peak_ind = np.argmax(M_region[:,1])
    return M_region[peak_ind]

```

```python
info = [[file[:-4], get_peak(file)] for file in files]
info
```

```python
ratio = info[0][1][1]/info[1][1][1]
print(r"Dynamical diffraction pi/sigma {:.3} \nCompared to {:.3f} using the $\cos(2\theta)^2$ rule".format(ratio, cos((110.06-90)*2)**2))
```
